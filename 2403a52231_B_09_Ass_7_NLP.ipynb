### Text Preprocessing Functions

```python
# Import necessary libraries
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download NLTK data if not already downloaded
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')
try:
    word_tokenize('test')
except LookupError:
    nltk.download('punkt')
try:
    WordNetLemmatizer().lemmatize('test')
except LookupError:
    nltk.download('wordnet')
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')
```

```python
def to_lowercase(text):
    """Converts text to lowercase."""
    return text.lower()

def remove_punctuation_numbers(text):
    """Removes punctuation and numbers from text."""
    # Create a translation table to remove punctuation
    translator = str.maketrans('', '', string.punctuation + string.digits)
    return text.translate(translator)

def remove_stopwords_func(tokens):
    """Removes English stopwords from a list of tokens."""
    stop_words = set(stopwords.words('english'))
    return [word for word in tokens if word not in stop_words]

def tokenize_text(text):
    """Tokenizes text into words."""
    return word_tokenize(text)

def lemmatize_tokens(tokens):
    """Lemmatizes a list of tokens."""
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word) for word in tokens]

def preprocess_text(text, lemmatize=False):
    """Applies a series of preprocessing steps to the text."""
    text = to_lowercase(text)
    text = remove_punctuation_numbers(text)
    tokens = tokenize_text(text)
    tokens = remove_stopwords_func(tokens)
    if lemmatize:
        tokens = lemmatize_tokens(tokens)
    return tokens
```

 TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Collect all documents into a list
documents = [
    d1, d2, d3, d4, d5, d6, d7, d8, d9, d10,
    d11, d12, d13, d14, d15, d16, d17, d18, d19, d20
]

# Preprocess each document using the previously defined function
# Join tokens back into a string, as TfidfVectorizer expects strings
preprocessed_documents = [" ".join(preprocess_text(doc, lemmatize=True)) for doc in documents]

# Initialize TF-IDF Vectorizer
# max_features can be set to limit the vocabulary size if needed
tfidf_vectorizer = TfidfVectorizer()

# Construct the TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(preprocessed_documents)

# Convert to DataFrame for better readability (optional)
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out(),
    index=[f'Document {i+1}' for i in range(len(documents))]
)

print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
print("\nSample of TF-IDF Matrix (first 5 documents, first 10 features):\n")
display(tfidf_df.head())
```

Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity between all documents
# tfidf_matrix is already computed and contains the TF-IDF vectors for each document
cosine_sim_matrix = cosine_similarity(tfidf_matrix)

# Convert to DataFrame for better readability, using original document indices
cosine_sim_df = pd.DataFrame(
    cosine_sim_matrix,
    index=[f'Document {i+1}' for i in range(len(documents))],
    columns=[f'Document {i+1}' for i in range(len(documents))]
)

print("Cosine Similarity Matrix:")
display(cosine_sim_df)
```

Jaccard Similarity

```python
def jaccard_similarity(doc1_tokens, doc2_tokens):
    """Calculates the Jaccard Similarity between two lists of tokens."""
    set1 = set(doc1_tokens)
    set2 = set(doc2_tokens)
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    if union == 0:
        return 0.0  # Handle cases where both sets are empty
    return intersection / union

# Compute Jaccard Similarity for all document pairs
num_documents = len(preprocessed_documents)
jaccard_sim_matrix = []

for i in range(num_documents):
    row = []
    for j in range(num_documents):
        sim = jaccard_similarity(preprocessed_documents[i], preprocessed_documents[j])
        row.append(sim)
    jaccard_sim_matrix.append(row)

# Convert to DataFrame for better readability
jaccard_sim_df = pd.DataFrame(
    jaccard_sim_matrix,
    index=[f'Document {i+1}' for i in range(num_documents)],
    columns=[f'Document {i+1}' for i in range(num_documents)]
)

print("Jaccard Similarity Matrix:")
display(jaccard_sim_df)
```

WordNet-based Semantic Similarity

```python
import nltk
from nltk.corpus import wordnet

# The `preprocess_text` function returns a list of tokens. We will use this directly.
tokenized_lemmatized_documents = [preprocess_text(doc, lemmatize=True) for doc in documents]

def wordnet_sentence_similarity(sentence1_tokens, sentence2_tokens, similarity_measure=wordnet.path_similarity):
    """
    Computes semantic similarity between two sentences using WordNet.
    Takes lists of preprocessed (tokenized, lemmatized) words.
    It averages the highest similarity score for each word in sentence1 with any word in sentence2.
    """
    if not sentence1_tokens or not sentence2_tokens:
        return 0.0

    # Filter out empty tokens which might arise from aggressive preprocessing
    sentence1_tokens = [token for token in sentence1_tokens if token]
    sentence2_tokens = [token for token in sentence2_tokens if token]

    if not sentence1_tokens or not sentence2_tokens:
        return 0.0

    overall_similarities = []

    for word1 in sentence1_tokens:
        # Try to get synsets for nouns and then verbs if no nouns found
        synsets1 = wordnet.synsets(word1, pos=wordnet.NOUN)
        if not synsets1:
            synsets1 = wordnet.synsets(word1, pos=wordnet.VERB)
        if not synsets1:
            continue

        max_sim_for_word1 = 0.0
        for word2 in sentence2_tokens:
            synsets2 = wordnet.synsets(word2, pos=wordnet.NOUN)
            if not synsets2:
                synsets2 = wordnet.synsets(word2, pos=wordnet.VERB)
            if not synsets2:
                continue

            # Compare the first (most common) synset of each word
            s1 = synsets1[0]
            s2 = synsets2[0]

            sim = similarity_measure(s1, s2)
            if sim is not None:
                max_sim_for_word1 = max(max_sim_for_word1, sim)

        # Only add if a similarity was found for word1
        if max_sim_for_word1 > 0:
            overall_similarities.append(max_sim_for_word1)

    if not overall_similarities:
        return 0.0
    return sum(overall_similarities) / len(overall_similarities)


# Select at least 10 sentence pairs for analysis
sentence_pairs_for_wordnet_analysis = [
    (0, 11),  # d1: 'I like machine learning', d12: 'Statistics is important for machine learning'
    (1, 6),   # d2: 'Artificial intelligence is very interesting', d7: 'Neural networks learn patterns from data'
    (8, 9),   # d9: 'Supervised learning uses labeled data', d10: 'Unsupervised learning finds hidden patterns'
    (3, 12),  # d4: 'Deep learning models improve accuracy', d13: 'Feature engineering improves model performance'
    (15, 16), # d16: 'Classification predicts categories', d17: 'Regression predicts continuous values'
    (0, 1),   # d1: 'I like machine learning', d2: 'Artificial intelligence is very interesting'
    (2, 18),  # d3: 'Data science helps in decision making', d19: 'Data visualization helps understand insights'
    (6, 7),   # d7: 'Neural networks learn patterns from data', d8: 'Big data requires powerful algorithms'
    (1, 4),   # d2: 'Artificial intelligence is very interesting', d5: 'Natural language processing is amazing'
    (13, 14)  # d14: 'Training data should be clean and balanced', d15: 'Model evaluation measures prediction quality'
]

wordnet_results = []
for idx1, idx2 in sentence_pairs_for_wordnet_analysis:
    doc1_tokens = tokenized_lemmatized_documents[idx1]
    doc2_tokens = tokenized_lemmatized_documents[idx2]
    sim = wordnet_sentence_similarity(doc1_tokens, doc2_tokens)
    wordnet_results.append({
        'Document 1': f'{documents[idx1]}',
        'Document 2': f'{documents[idx2]}',
        'WordNet Similarity (Path)': sim
    })

wordnet_results_df = pd.DataFrame(wordnet_results)
print("WordNet Semantic Similarity for Selected Pairs (using Path Similarity):")
display(wordnet_results_df)
```
