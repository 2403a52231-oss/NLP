1 question:

```python
Report="""She used to have allergies when she lived in Seattle but she thinks they are worse here.  In the past, she has tried.,A 42-year-old male presents with a persistent dry cough and "tightness" in the chest during physical activity. He has no childhood history of respiratory disease but moved to a city with high industrial pollution two years ago.,he patient is a 34-year-old female reporting a significant increase in allergic rhinitis symptoms since relocating from Seattle to a drier, high-pollen region. She notes that her previous baseline of mild seasonal congestion has progressed to daily distress.A 29-year-old female presents with "crushing fatigue" and muscle weakness that has persisted for six months. She recently transitioned to a strict vegan diet and works in a windowless office environment."""
```

```python
print(Report)
```

Tokenizing word

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
word_tokenize(Report)
```

Tokenizing by sentence

```python
from nltk.tokenize import sent_tokenize
sent_tokenize(Report)
```

Stop Words

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
words_in_quote = word_tokenize(Report)
words_in_quote
```

```python
stop_words = set(stopwords.words("english"))
filtered_list = []
for word in words_in_quote:
  if word.casefold() not in stop_words:
    filtered_list.append(word)
filtered_list
```

Stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
stemmer = PorterStemmer()
words = word_tokenize(Report)
stemmed_words = [stemmer.stem(word) for word in words]
stemmed_words
```

Snowball stemmer

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer(language='english')
words = word_tokenize(Report)
for word in words:
    print(word,"--->",snowball.stem(word))
```

```python
from nltk import LancasterStemmer
Lanc = LancasterStemmer()
words = word_tokenize(Report)
for word in words:
    print(word,"--->",Lanc.stem(word))
```

```python
from nltk.stem import RegexpStemmer
regexp = RegexpStemmer('ing|s|e|able', min=4)
words = word_tokenize(Report)
for word in words:
    print(word,"--->",Lanc.stem(word))
```

Lemmatization

```python
import nltk
nltk.download('omw-1.4')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
words = word_tokenize(Report)
for word in words:
    print(word,"--->",lemmatizer.lemmatize(word))
```

Comparison

```python
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer, RegexpStemmer, WordNetLemmatizer
porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer(language='english')
regexp = RegexpStemmer('ing|s|e|able', min=4)
lemmatizer = WordNetLemmatizer()

word_list = ["friend", "friendship", "friends", "friendships"]
print("{0:20}{1:20}{2:20}{3:30}{4:40}{5:50}".format("Word","Porter Stemmer","Snowball Stemmer","Lancaster Stemmer",'Regexp Stemmer','WordNetLemmatizer'))
for word in word_list:
    print("{0:20}{1:20}{2:20}{3:30}{4:40}{5:50}".format(word,porter.stem(word),snowball.stem(word),lancaster.stem(word),regexp.stem(word),lemmatizer.lemmatize(word)))
```
