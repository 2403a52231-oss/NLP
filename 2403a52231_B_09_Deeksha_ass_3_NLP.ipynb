```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
Sentence = "Student are learning Natural Language Processing"
tokens = nltk.word_tokenize(Sentence)
pos_tags = nltk.pos_tag(tokens)
print(pos_tags)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(Sentence)
for token in doc:
    print(token.text, token.pos_)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying a startup in India")
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
import spacy
import collections
from collections import Counter
nlp = spacy.load("en_core_web_sm")
text = "Loving the new AI features ğŸ˜ #AI #MachineLearning"
doc = nlp(text)
nouns = []
verbs = []
for token in doc:
    if token.pos_ == "NOUN":
        nouns.append(token.text)
    elif token.pos_ == "VERB":
        verbs.append(token.text)
noun_freq = Counter(nouns)
verb_freq = Counter(verbs)
print("Nouns:", noun_freq)
print("Verbs:", verb_freq)
```
